Δημιουργούμε έναν πίνακα με το link, τον τίτλο, την ημερομηνία, την περίληψη και την εισαγωγή κάθε άρθρου.

```python
def getUrlAndCounts():
    df = pd.DataFrame()
    rows = doc.find_all('article', attrs={'class': 'teaser__article'})
    for row in rows:
      try:
        link = row.find('a')['href']
        title = row.find('a') ['title']
        date = row.find('span', attrs={'class':'article__time'}).text
        summary = row.find('div', attrs={'class':'article__summary'}).text
        intro = row.find('p', attrs={'class':'article__leadtext'}).text

        data ={
              'Link': link,
              'Title': title,
              'Date': date,
              'Summary': summary,
              'Intro': intro
          }
        df = df.append(data, ignore_index=True)  
      except:
        print("skip") 
    return df    

rows = getUrlAndCounts()
rows.tail(10)
```

Δημιουργούμε μια στήλη ακόμη στον πίνακα που περιέχει ένα τμήμα του κειμένου από κάθε άρθρο

```python
from requests.models import Response
def getUrlAndCounts():
    df = pd.DataFrame()
    rows = doc.find_all('article', attrs={'class': 'teaser__article'})
    for row in rows:
      try:
          link = row.find('a')['href']
          response = requests.get(link)
          doclink = BeautifulSoup(response.text, 'html.parser')
          text = doclink.find('div', {'class': 'article-body__body'}).text
          title = row.find('a') ['title']
          date = row.find('time').text
          summary = row.find('p', attrs={'class':'article__leadtext'}).text
          tag = row.find('a',{"class": "article__tag_section"}).text
          
            
          print(text)
          data ={
              'Link': link,
              'Title': title,
              'Date': date,
              'Summary': summary,
              'Text' : text,
              'Tag' : tag
          }
          df = df.append(data, ignore_index=True)
      except:
        print("skip") 
    return df
 
rows = getUrlAndCounts()
rows.tail(10)
```
